```python
def b_sort(k, s):
    if k is not None:
        for i in range(s-1):
            for j in range(s-1-i):
                if k[j] > k[j+1]:
                    temp = k[j]
                    k[j] = k[j+1]
                    k[j+1] = temp
        return k

def li():
    k = []
    while True:
        i = int(input())
        if i == 1000:
            break
        k.append(i)
    return k, len(k)

m, s = li()
m = b_sort(m, s)
if m:
    print(m)
else:
    print('is empty')
```
