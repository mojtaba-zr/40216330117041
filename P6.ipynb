```python
def bubble_sort(k=[], s=10):
    if k is not None:
       for i in range (s-1):
          for j in range (s-1-i):
             if k[j] > k[j+1]:
                temp = k[j]
                k[j] = k[j+1]
                k[j+1] = temp
                return 1
    else:
       return 0

l = [3, 1, 9, 77, 21, 2]
s = bubble_sort(l, 6)
print(s)
```
