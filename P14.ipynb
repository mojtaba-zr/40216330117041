```python
import math

class Circle:
  def __init__(self, shoa):
    self.r = shoa
    self.pi = math.pi
    print("Circle created.")

  def mohit(self):
    mo = 2 * self.pi * self.r
    return mo

  def masahat(self):
    ma = self.pi * self.r * self.r
    return ma

  def __del__(self):
    print("Circle deleted.")

d = Circle(4)
print(d.mohit())
print(d.masahat())
del d
```
