```python
class students:
    def __init__(self, id, name, grades):
        self.id = id
        self.name = name
        self.grades = grades

    def __del__(self):
        print(f"{self.name} Deleted")

    def GPA(self):
        return sum(self.grades) / len(self.grades)

    def conditional_grade_point_average(self):
        if self.GPA() < 12:
            return True
        else:
            return False

    def __str__(self):
        return f"Student ID: {self.id}, Name: {self.name}, GPA: {self.GPA():.2f}"

student1 = students(1, "John", [10, 12, 11, 13])
print(student1.GPA())
print(student1.conditional_grade_point_average())
print(student1)
```
